```python
import pandas as pd
```

```python
headers = ["pitch", "roll"]
sources = ["10", "20", "30", "40", "50", "60", "70", "80"]
dfs = []
for source in sources:
    data = pd.read_csv(f'prototype_data/blocks_{source}.csv', names=headers)
    dfs.append(data)
    
roll_data = pd.read_csv('prototype_data/blocks_roll_90.csv', names=headers)
```

```python
variance_results = []

for source, df in zip(sources, dfs):
    pitch_variance = df['pitch'].var()
    pitch_min = df['pitch'].min()
    pitch_max = df['pitch'].max()
    roll_variance = df['roll'].var()
    roll_min = df['roll'].min()
    roll_max = df['roll'].max()
    datapoints = df.shape[0]
    variance_results.append({'source': source, 'pitch_variance': pitch_variance, 'pitch_min': pitch_min, 'pitch_max': pitch_max, 
                             'roll_variance': roll_variance, 'roll_min': roll_min, 'roll_max': roll_max, 'datapoints': datapoints})

pitch_variance = roll_data['pitch'].var()
pitch_min = roll_data['pitch'].min()
pitch_max = roll_data['pitch'].max()
roll_variance = roll_data['roll'].var()
roll_min = roll_data['roll'].min()
roll_max = roll_data['roll'].max()
datapoints = roll_data.shape[0]
variance_results.append({'source': 'roll_90', 'pitch_variance': pitch_variance, 'pitch_min': pitch_min, 'pitch_max': pitch_max, 
                             'roll_variance': roll_variance, 'roll_min': roll_min, 'roll_max': roll_max, 'datapoints': datapoints})

variance_df = pd.DataFrame(variance_results)
display(variance_df)

avg_df = pd.DataFrame([{'avg_variance_pitch': variance_df['pitch_variance'].mean(), 'avg_variance_roll': variance_df['roll_variance'].mean(), 'datapoints': variance_df['datapoints'].sum()}])
display(avg_df)
```
